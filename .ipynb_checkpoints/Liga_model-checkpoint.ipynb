# Fencing bouts prediction
## Comparisson of classification methods
In this notebook I will attempt to create a model that predicts outcomes of fencing bouts fought between athleets from Krakowska Szkoła Fechtunku. Data used to train the model is a combination of data genereted during figts in clubs internal league and fencers metrics gathered by me.
After combining the datasets, and performing an EDA I will train 3 models: logistic regression, KNN, and support vector machines, evaluate them and choose the best one.
The last step is creating the function that generates pairs that will fight during next league session, and automaticly makes prediction for this pairs.

```python
#This function checks data frame for missing values and returns sorted series of columns with respective 
#% of missing data, and shows theese values on plot. If there is no missing values it prompts it.
def p_missing(df):
    percent = 100*df.isnull().sum()/len(df)
    percent = percent[percent>0].sort_values()
    if df.isnull().values.sum() > 0:
        plot = sns.barplot(x=percent.index,y=percent)
        plt.xticks(rotation=90);
    else:
        return print('No cols with missing values')
        return plot, percent
```

```python
#Inserting fencers metrics age, weight, and height
def ins(df, fencer_num, name, age, weight, height, main_weapon, yrs_of_training, comp):
    fencer_col = f'Fencer {fencer_num}'
    df.loc[df[fencer_col]==name, f'F{fencer_num} age'] = age
    df.loc[df[fencer_col]==name, f'F{fencer_num} bodyweight'] = weight
    df.loc[df[fencer_col]==name, f'F{fencer_num} height'] = height
    df.loc[df[fencer_col]==name, f'F{fencer_num} main weapon'] = main_weapon
    df.loc[df[fencer_col]==name, f'F{fencer_num} yrs of training'] = yrs_of_training
    df.loc[df[fencer_col]==name, f'F{fencer_num} yrs of training as competitor'] = comp
    return df

d={'Michalina':[27, 62, 163, 'Miecz',4,1],
  'Wojtek':[26,80,180, 'Miecz',3,1],
  'Michał':[30, 75,178, 'Miecz',3,1],
  'Kumiś':[38,85,185, 'Miecz',10,4],
  'Kuba':[32,80,183, 'Miecz',4,1],
  'Franek':[25,84,187, 'Szabla',10,5],
  'Łukasz':[34,73,176, 'Miecz',10,2],
  'Adzik':[31,62,163, 'Miecz',7,5],
  'Jurek':[36,90,190, 'Miecz',16,0],
  'Jacek':[25,178,110, 'Miecz',10,5],
  'Dawid':[33,180,83, 'Miecz',8,5]}
for k,v in d.items():
    data = ins(data, 1 ,k, v[0],v[1],v[2], v[3],v[4],v[5])
    data = ins(data, 2 ,k, v[0],v[1],v[2], v[3],v[4],v[5])
    print(k)
```

```python
data['F1 lifts']=data['F1 squat max']+data['F1 deadlift max']+data['F1 benchpress max']
data['F2 lifts']=data['F2 squat max']+data['F2 deadlift max']+data['F2 benchpress max']
data[['Fencer 1','F1 lifts','F1 squat max','F1 deadlift max', 'F1 benchpress max']]
```

## Data preprocessing

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
#Performing train test split
X=final_df.drop('F1_win?', axis=1)
y=final_df['F1_win?']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
#encoding categorical columns
s = X_train.dtypes == object
object_cols = list(s[s].index)
enc = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
t_X_train=enc.fit_transform(X_train[object_cols])
t_X_test=enc.transform(X_test[object_cols])
feature_names = enc.get_feature_names_out(input_features=object_cols)
t_X_train = pd.DataFrame(t_X_train, columns=feature_names)
t_X_test = pd.DataFrame(t_X_test, columns=feature_names)
t_X_train.index = X_train.index
t_X_test.index = X_test.index
```

```python
#Combining cumerical and categorical features back into one df
num_X_train = X_train.drop(object_cols, axis=1)
num_X_test = X_test.drop(object_cols, axis=1)
OH_X_train = pd.concat([num_X_train,t_X_train], axis=1)
OH_X_test = pd.concat([num_X_test,t_X_test], axis=1)
```

```python
#scaling data
scaler=StandardScaler()
X_train=scaler.fit_transform(OH_X_train)
X_test=scaler.transform(OH_X_test)
```

## Training models


### Logistic regression

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
log_model = LogisticRegressionCV(Cs = np.logspace(0,10,20),
                               solver='saga',
                               multi_class="ovr",
                               max_iter=5000,
                               penalty='l1')
log_model.fit(X_train, y_train)
```

```python
log_model.C_
```

```python
log_model.get_params()
```

```python
log_model.coef_
```

```python
coefs = pd.Series(index=OH_X_train.columns,data=log_model.coef_[0])
coefs = coefs.sort_values()
plt.figure(figsize=(10,6))
sns.barplot(x=coefs.index,y=coefs.values)
plt.xticks(rotation=90);
```

As per plot above the only relevant features are F1, and F2 touches. Coeficents of other features were shrinked to very small values.

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,ConfusionMatrixDisplay,roc_curve,RocCurveDisplay,auc
```

```python
y_pred = log_model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

The model achieved perfect scores on test set. it looks kinda suspicious, but on the other hand the problem was trivial. It is extremely easy to predict the outcome of a bout when you already know the scores. In that context such performance on the test set is a good sign and was expected. The real test will be the prediction of future bouts. 


### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
k_val = list(range(1,51))
param_grid = dict(n_neighbors=k_val)
grid = GridSearchCV(knn, param_grid, cv=10, scoring='accuracy')
```

```python
grid_search=grid.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_.get_params()
```

```python
scores = grid_search.cv_results_['mean_test_score']
plt.plot(k_val,scores,'-o')
plt.xlabel("K")
plt.ylabel("Accuracy")
```

```python
pred = grid_search.predict(X_test)
print(confusion_matrix(pred,y_test))
print(classification_report(pred,y_test))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,pred)
```

KNN also performed very good, but it produced two false positives.

## Predicting future bouts

This section is the true purpose of creating these models. As indicated earlier predicting the outcomes when you know them is easy. In this section we will try to predict the outcomes of the latest league seasion wich took place after the dataset had been completed. For that task the average num of touches for a given fencer during his newest league sesion will be used.

```python
#Creating data frame with average touches per league
df_f1=data[['Fencer 1','League number','F1 touches','Double hits','Elo_F1_before']]
df_f2=data[['Fencer 2','League number','F2 touches','Double hits','Elo_F2_before']]
```

```python
Fencers = [df_f1,df_f2]
new_column_names = ['Fencer','League number', 'Touches', 'Double hits', 'Elo']
for df in Fencers:
    df.columns = new_column_names
fencers_grouped = pd.concat([df_f1,df_f2])
fencers_grouped =fencers_grouped.groupby(['League number', 'Fencer']).mean()
fencers_grouped=fencers_grouped.reset_index()
fencers_pred = data
```

```python
# Inserting average touches into final_df
data = fencers_pred
for index, row in fencers_grouped.iterrows():
    fencers_pred['F1 touches'].loc[(fencers_pred['Fencer 1'] == row['Fencer']) & (fencers_pred['League number'] == row['League number'])]=row['Touches']
    fencers_pred['F2 touches'].loc[(fencers_pred['Fencer 2'] == row['Fencer']) & (fencers_pred['League number'] == row['League number'])]=row['Touches']
    fencers_pred['Elo_F1_before'].loc[(fencers_pred['Fencer 1'] == row['Fencer']) & (fencers_pred['League number'] == row['League number'])]=row['Elo']
    fencers_pred['Elo_F2_before'].loc[(fencers_pred['Fencer 2'] == row['Fencer']) & (fencers_pred['League number'] == row['League number'])]=row['Elo']
```

```python
#fropping target for pred
fencers_pred = fencers_pred.drop(['F1_win?'], axis=1)
```

```python
#encoding dummies
fencers_pred=pd.get_dummies(fencers_pred,dtype=int)
fencers_pred 
```
