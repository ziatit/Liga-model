```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_Miecz = pd.read_csv("History_Miecz.txt")
df_Miecz['Wygrany_ID'] = df_Miecz['Wygrany']
df_Miecz.rename(columns={'Wygrany':'Fencer 1',
                                  'Przegrany':'Fencer 2',
                                 'Trafienia W. Znorm.':'F1 touches',
                                 'Trafienia P. Znorm':'F2 touches',
                                 'Duble Znorm.':'Double hits',
                                 'Numer Ligi':'League number',
                                 'Elo_w_before':'Elo_F1_before',
                                 'Elo_l_before':'Elo_F2_before'
                                 }, inplace = True)
df_Miecz.head()
```

```python
for i in range(0, len(df_Miecz), 2):
    df_Miecz.at[i, 'Fencer 1'], df_Miecz.at[i, 'Fencer 2'], df_Miecz.at[i, 'F1 touches'], df_Miecz.at[i, 'F2 touches'], df_Miecz.at[i, 'Elo_w_after'], df_Miecz.at[i, 'Elo_l_after'], df_Miecz.at[i, 'Elo_F1_before'], df_Miecz.at[i, 'Elo_F2_before'] = df_Miecz.at[i, 'Fencer 2'], df_Miecz.at[i, 'Fencer 1'],df_Miecz.at[i, 'F2 touches'],df_Miecz.at[i, 'F1 touches'],df_Miecz.at[i, 'Elo_l_after'],df_Miecz.at[i, 'Elo_w_after'],df_Miecz.at[i, 'Elo_F2_before'],df_Miecz.at[i, 'Elo_F1_before']
```

```python
df_Miecz.drop(['Trafienia Wygranego', 'Trafienia Przegranego', 'Duble','Maks punktów' ], axis=1, inplace=True)
```

```python
X = data.drop('F1_win?',axis=1)
y = data['F1_win?']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV 
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
Cs = [0.001, 0.01, 0.1, 1, 10]
log_model = LogisticRegressionCV(Cs=Cs, random_state=101, max_iter=1000)
log_model.fit(X_train, y_train)
```

```python
log_model.C_
```

```python
log_model.get_params()
```

```python
coefs = coefs = pd.Series(index=X.columns,data=log_model.coef_[0])
coefs.sort_values()
```

```python
coefs = coefs.sort_values()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=coefs.index,y=coefs.values);
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,ConfusionMatrixDisplay,roc_curve,RocCurveDisplay,auc
```

```python
y_pred = log_model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
df_Miecz[['F1 touches', 'F2 touches', 'Double hits']]=df_Miecz[['F1 touches', 'F2 touches','Double hits']].apply(lambda x:x.str.replace(',','.')).astype(float)
df_F1 = df_Miecz[['Fencer 1','League number','F1 touches','Double hits','Elo_F1_before' ]]
df_F2 = df_Miecz[['Fencer 2', 'League number', 'F2 touches', 'Double hits', 'Elo_F2_before']]
```

```python
grouped_F1 = df_F1.groupby(['Fencer 1', 'League number']).mean().sort_values(by='League number', ascending=True)
```

```python
grouped_F1
```

```python
Fencers = [df_F1, df_F2]
new_column_names = ['Touches', 'Double hits', 'Elo']
for df in Fencers:
    df.columns = new_column_names
```

```python
Fencers_final = pd.concat([df_F2,df_F1])
```

```python
F_gr=Fencers_final.groupby(['League number', 'Fencer']).mean()
F_gr
```

```python
def make_preds(pairs, model):
    for key, value in pairs.items():
        l = F_gr.loc[(F_gr.index.max()[0], key)].tolist()    
        for v in F_gr.loc[(F_gr.index.max()[0], value)]:
            l.append(v)
        pred = [[0,0,0,0,0]]
        pred[0][0] = l[0]
        pred[0][1] = l[3]
        pred[0][2] = (l[1] + l[4]) / 2
        pred[0][3] = l[2]
        pred[0][4] = l[-1]
        result = model.predict(pred)
        key = str(key)
        value = str(value)
        if result == 1:
            print('Model predicts that in next fight '+ key +' will win with '+ value)
        else:
            print('Model predicts that in next fight '+ value +' will win with '+ key)
```

```python
pairs = {'Przemo':'Michał', 'Artem':'Michał'}
make_preds(pairs, log_model)
```
